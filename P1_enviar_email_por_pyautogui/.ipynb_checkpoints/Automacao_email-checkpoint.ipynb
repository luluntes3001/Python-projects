# Automação de envio de relatório por email utilizando a biblioteca pyautogui

**Cenário:**

Todos os dias, o sistema atualiza as vendas do dia anterior. Portanto, deve-se enviar um email para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos.

Email da diretoria: Exemplo.email@gmail.com
Local onde o sistema disponibiliza as vendas do dia anterior:linkdrive.

```python
# Importante: Lembrar de instalar a biblioteca do pyautogui
# Será realizado no Windows 10 no navegador google Chrome

import pyautogui
import pyperclip
import time
import pandas as pd

pyautogui.PAUSE = 1 #cada comando terá a pausa de 1 seg entre eles

# Passo 1: Entrar no sistema (neste caso vai ser o link do drive)

pyautogui.hotkey("ctrl", "t") 
pyautogui.write("Seu link do drive")

# Caso o link tivesse caracteres especiais é possível usar o paperclip para contornar esse problema

#pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga")
#pyautogui.hotkey("ctrl", "v")

pyautogui.press("enter")

#Em caso de internet lenta pode dar uma pausa para o página carregar
time.sleep(5)

# Passo 2: Navegar no sistema e encontrar a base de dados (entrar na pasta Exportar)
pyautogui.click(x=351, y=274, clicks=2)
time.sleep(5)

# Passo 3: Download da base de dados
pyautogui.click(x=384, y=279)
pyautogui.click(x=1139, y=171)
time.sleep(2)
pyautogui.click(x=909, y=645)
time.sleep(5)
pyautogui.press("enter") #caso abra a janela para escolher o diretório
time.sleep(10)
```

```python
# Passo 4: Calcular indicadores (faturamento, quantidade de produtos)
tabela = pd.read_excel(r"SEU DIRETÓRIO")
display(tabela)
quantidadeProd = tabela["Quantidade"].sum()
faturamento = tabela["Valor Final"].sum()

# Passo 5: Entrar no email (aqui no caso será o gmail)
pyautogui.hotkey("ctrl", "t") 
pyautogui.write("https://mail.google.com/mail/u/0/#inbox")
pyautogui.press("enter")
time.sleep(15)

# Passo 6: Mandar um email para a diretoria com os indicadores
pyautogui.click(x=65, y=180)
time.sleep(10)
pyautogui.write("SEUEMAIL")
pyautogui.press("tab")
pyautogui.press("tab")
time.sleep(2)
pyautogui.write("Relatorio de vendas")
pyautogui.press("tab")
time.sleep(2)

# Passo 6.1: Escrever o email e enviar
texto = f"""Prezados, bom dia

O faturamento de ontem foi de: R$ {faturamento:,.2f}
A quantidade de produtos vendidos foi de: {quantidadeProd:,}

Abs
"""
pyautogui.write(texto)
time.sleep(2)
pyautogui.hotkey("ctrl", "enter")

#Fim
```
